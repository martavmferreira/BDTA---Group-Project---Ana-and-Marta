```python
# Calculation of the engagement rate (total likes divided by followers) on Instagram for each festival 
# and saving it to a CSV file

import pandas as pd

df_likes = pd.read_csv("festivals_instagram_likes.csv")  
df_followers = pd.read_csv("festivals_instagram_followers.csv") 

# Calculate total likes and total Followers for each festival
total_likes = df_likes.groupby("Festival")["Likes"].sum()
total_followers = df_followers.groupby("Festival")["Followers"].sum()

# Calculate the engagement rate (likes per follower)
engagement_rate_instagram = (total_likes / total_followers) * 100

# Save it to a CSV file
engagement_rate_instagram.to_csv("festivals_instagram_engagement.csv", header=['Engagement Rate'], index_label='Festival')
```

```python
# Calculation of the engagement rate (total likes divided by followers) on Twitter for each festival 
# and saving it to a CSV file

import pandas as pd

# Load the Twitter likes and followers data
df_likes = pd.read_csv("festivals_twitter_likes.csv")  # Replace with your actual filename
df_followers = pd.read_csv("festivals_twitter_followers.csv")  # Replace with your actual filename

# Calculate Total Likes and Total Followers for Each Festival
total_likes = df_likes.groupby("Festival")["Likes"].sum()
total_followers = df_followers.groupby("Festival")["Followers"].sum()

# Calculate Engagement Rate (Likes per Follower)
engagement_rate_twitter = (total_likes / total_followers) * 100

# Save the engagement rate data to a CSV file
engagement_rate_twitter.to_csv("festivals_twitter_engagement.csv", header=['Engagement Rate'], index_label='Festival')
```

```python
# Likes on Instagram

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the Instagram likes data
df_likes = pd.read_csv("festivals_instagram_likes.csv")

# Set the Festival column as the index
df_likes.set_index("Festival", inplace=True)

# Sort the dataframe by Likes in descending order
df_likes_sorted = df_likes.sort_values(by="Likes", ascending=False)

palette = sns.color_palette('pastel')
colors = [palette[1], palette[2], palette[0], palette[3], palette[4]]

# Plotting
plt.figure(figsize=(10, 6))
df_likes_sorted["Likes"].plot(kind="bar", color=colors)
plt.title("Total Instagram Likes per Festival")
plt.xlabel("Festival")
plt.ylabel("Number of Likes")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
# Engagement Rate for Instagram

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Twitter likes and followers data
df_likes = pd.read_csv("festivals_instagram_likes.csv")  # Replace with your actual filename
df_followers = pd.read_csv("festivals_instagram_followers.csv")  # Replace with your actual filename

# Step 1: Calculate Total Likes and Total Followers for Each Festival
total_likes = df_likes.groupby("Festival")["Likes"].sum()
total_followers = df_followers.groupby("Festival")["Followers"].sum()

# Step 2: Calculate Engagement Rate (Likes per Follower)
engagement_rate = (total_likes / total_followers) * 100

# Sort the engagement rate from highest to lowest
engagement_rate_sorted = engagement_rate.sort_values(ascending=False)

palette = sns.color_palette('pastel')
colors = [palette[1], palette[2], palette[4], palette[3], palette[0]]

# Step 3: Visualize the Engagement Rate
plt.figure(figsize=(10, 6))
sns.barplot(x=engagement_rate_sorted.index, y=engagement_rate_sorted.values, palette=colors)
plt.title("Engagement Rate on Instagram per Festival")
plt.xlabel("Festival")
plt.ylabel("Engagement Rate (%)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```
