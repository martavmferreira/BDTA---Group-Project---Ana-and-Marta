```python
import pandas as pd
df= pd.read_csv("artists popularity data.csv", low_memory = False)
df.head()
```

```python
columns_to_remove = ['mbid', 'artist_mb', 'country_mb','tags_mb']  # List of column names to remove
df_cleaned = df.drop(columns=columns_to_remove)
df_cleaned.head(10)
```

```python
# Sort the DataFrame by descending order
df_sorted = df_cleaned.sort_values(by=['listeners_lastfm'], ascending=False)
df_sorted.head(20)
```

```python
# Create a dictionary with the headliners of each festival
data = {
    'Festival': ['Rock in Rio', 'NOS Alive', 'MEO Sudoeste', 'Super Bock Super Rock', 'Vodafone Paredes de Coura'],
    'Artists': [
        ['Scorpions', 'Ed Sheeran', 'Lukas Graham', 'Jonas Brothers', 'Doja Cat'],
        ['Arcade Fire', 'The Smashing Pumpkins', 'Dua Lipa', 'Pearl Jam', 'Floating Points'],
        ['Martin Garrix', 'Anitta', 'Charlotte de Witte', 'Richie Campbell', 'Da Weasel'],
        ['21 Savage', 'Fisher', 'Royal Blood', 'Mahalia', 'Vulfpeck'],
        ['Dorian Concept', 'Superchunk', 'Sampha', 'Los Bitchos', 'Idles']
    ]
}

# Store the data inside a DataFrame
Headliners = pd.DataFrame(data)

Headliners.head()
```

```python
# Create a list with all Headliners
all_headliners = [artist for artists_list in Headliners['Artists'] for artist in artists_list]

# Filter the artists present in the DataFrame df_cleaned
filtered_artists = df_cleaned[df_cleaned['artist_lastfm'].isin(all_headliners)]

filtered_artists.head(5)
```

```python
import numpy as np

# Calculate the average of listeners and scrobbles for artists with ambiguity
ambiguous_artist = filtered_artists[filtered_artists['ambiguous_artist'] == True]
ambiguous_artist_mean = ambiguous_artist.groupby('artist_lastfm').agg({'listeners_lastfm': 'mean', 'scrobbles_lastfm': 'mean'}).reset_index()

# Remove duplicates from ambiguous artists
filtered_artists = filtered_artists.drop_duplicates(subset=['artist_lastfm'])

# Fill in the average of listeners and scrobbles for ambiguous artists
filtered_artists = filtered_artists.merge(ambiguous_artist_mean, on='artist_lastfm', suffixes=('', '_mean'), how='left')
filtered_artists['listeners_lastfm'] = filtered_artists['listeners_lastfm_mean'].fillna(filtered_artists['listeners_lastfm'])
filtered_artists['scrobbles_lastfm'] = filtered_artists['scrobbles_lastfm_mean'].fillna(filtered_artists['scrobbles_lastfm'])

# Keep relevant columns
filtered_artists = filtered_artists.drop(['listeners_lastfm_mean', 'scrobbles_lastfm_mean'], axis=1)

# Add 'Festival' column to the filtered Dataframe 
for index, row in Headliners.iterrows():
    festival_name = row['Festival']
    filtered_artists.loc[filtered_artists['artist_lastfm'].isin(row['Artists']), 'Festival'] = festival_name

filtered_artists.head(1000)
```

```python
# Group the artists by festival and calculate the average number of listeners
avg_listeners_by_artist = filtered_artists.groupby(['Festival', 'artist_lastfm'])['listeners_lastfm'].mean().reset_index()

# Sort the artists by the number of listeners in descending order
sorted_artists = avg_listeners_by_artist.sort_values(by='listeners_lastfm', ascending=False)

sorted_artists[['artist_lastfm', 'listeners_lastfm', 'Festival']].head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.cm as cm
import matplotlib.patches as mpatches

# Group the data by festival and artists + calculate the listeners average
grouped_artists = filtered_artists.groupby(['Festival', 'artist_lastfm'])['listeners_lastfm'].mean().reset_index()

# Sort festivals by listeners average for each festival
sorted_festivals = grouped_artists.groupby('Festival')['listeners_lastfm'].mean().sort_values(ascending=False).index

# Create a data matrix for stacked bars
data_matrix = np.zeros((len(grouped_artists['artist_lastfm'].unique()), len(sorted_festivals)))

# Generate unique colours for each artist
num_artists = len(grouped_artists['artist_lastfm'].unique())
colors_tab20b = [plt.cm.tab20b(i / 20) for i in range(20)]
colors_tab20c = [plt.cm.tab20c(i / 20) for i in range(20)]
colors = colors_tab20b + colors_tab20c  # Combining colours from different colormaps

plt.figure(figsize=(12, 8))

legend_patches = []  # List to store Patch objects for the legend

for i, artist in enumerate(grouped_artists['artist_lastfm'].unique()):
    artist_data = grouped_artists[grouped_artists['artist_lastfm'] == artist].groupby('Festival')['listeners_lastfm'].mean()
    for j, festival in enumerate(sorted_festivals):
        data_matrix[i, j] = artist_data.get(festival, 0)
        plt.bar(sorted_festivals, data_matrix[i], label=None, bottom=np.sum(data_matrix[:i], axis=0), color=colors[i])
    
    # Add the patch object to the legend
    legend_patches.append(mpatches.Patch(color=colors[i], label=artist))

plt.xticks(rotation=45, ha='right')
plt.xlabel('Festival')
plt.ylabel('Average Listeners')
plt.title('Average Listeners per Artist and Festival')

# Create the legend with the Patch objects
plt.legend(handles=legend_patches, loc='upper center', bbox_to_anchor=(0.5, -0.4), fontsize='small', ncol=5)

plt.subplots_adjust(bottom=0.3)

# Exibir o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a dataframe with the average number of listeners for each festival
mean_listeners_df = filtered_artists.groupby('Festival')['listeners_lastfm'].mean().reset_index()

# Set up the data for the radar chart
labels = mean_listeners_df['Festival']
stats = mean_listeners_df['listeners_lastfm']
stats = np.concatenate((stats, [stats[0]]))  # Add the first value at the end to close the chart

angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], labels, color='grey', size=8, rotation=45, ha='right')  # Ajustar os rótulos dos festivais
ax.plot(angles, stats, linewidth=1, linestyle='solid', label='Average Listeners', color='lightblue')  # Definir a cor como azul clara
ax.fill(angles, stats, 'lightblue', alpha=0.1)  # Preencher a área com a mesma cor
plt.title('Average Listeners by Festival')
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```
