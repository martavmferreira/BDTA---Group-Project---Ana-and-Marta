```python
import folium
import matplotlib.pyplot as plt

festivals = {
    'NOS ALIVE': (38.69840499127016, -9.23810536246993),
    'ROCK IN RIO': (38.78566263943525, -9.092905520136062),
    'MEO SUDOESTE': (37.55441883342442, -8.731169743495334),
    'SUPER BOCK SUPER ROCK': (38.52589461321114, -9.146853563335743),
    'VODAFONE PAREDES DE COURA': (41.95845113232415, -8.536732035071669)
}

# Create a map using folium
map = folium.Map(location=[39.3999, -8.2245], zoom_start=8)

# Add markers where the festivals are
for festival, coord in festivals.items():
    folium.Marker(location=coord, popup=festival).add_to(map)

map.save('festivals_map.html')
map
```
