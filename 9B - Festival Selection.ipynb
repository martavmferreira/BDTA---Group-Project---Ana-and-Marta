```python
import pandas as pd

normalized_data = (r"C:\Users\a1927\Desktop\MESTRADO\Semester 2\BIG DATA\Group project\festival selection.xlsx")

selection = pd.read_excel(normalized_data)
selection.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

palette = sns.color_palette('pastel')

# Show the best festival until the "worst"
selection_sorted = selection.sort_values(by='TOTAL', ascending=False)

# Create the graph:
plt.figure(figsize=(10, 6))
plt.barh(selection_sorted['MUSIC EVENTS'], selection_sorted['TOTAL'], color=palette)

plt.title('Festival Ranking')
plt.xlabel('Total Score')
plt.ylabel('Festival')

plt.gca().invert_yaxis()  
plt.grid(False)  
plt.tight_layout()
plt.show()
```
