```python
# Create a new column in NOS Alive CSV file
# to categorize the type of ticket 'Daily','Two Days Pass','All Days Pass' or 'Other'
import pandas as pd

# Load the ticket prices data for the specific festival into a DataFrame
ticket_prices_alive = pd.read_csv("ticket_prices_alive.csv")

# Function to categorize ticket type for the specific festival
def categorize_ticket_type(ticket_name):
    ticket_name_lower = ticket_name.lower()
    
    if any(word in ticket_name for word in ['ENTRADA', 'DIÁRIO', 'JULY', 'AUGUST' 'August']) and 'VIP' not in ticket_name:
        return 'Daily'
    elif any(word in ticket_name for word in ['2 DIAS', 'ABONO', 'FIM DE SEMANA', 'PASS WITHOUT']):
        return 'Two Days Pass'
    elif any(word in ticket_name for word in ['3 DIAS', 'Pass', 'GENERAL PASS WITHOUT', 'PASS WITHOUT', 'Full']):
        return 'All Days Pass'
    else:
        return 'Other'

# Add a new column indicating ticket type for the specific festival
ticket_prices_alive['Ticket Category'] = ticket_prices_alive['Tipo de Bilhete'].apply(categorize_ticket_type)

# Save the DataFrame with the modified column to a new CSV file
ticket_prices_alive.to_csv("ticket_prices_alive_categorized.csv", index=False)

ticket_prices_alive
```

```python
# Create a new column in Rock in Rio CSV file
# to categorize the type of ticket 'Daily','Two Days Pass','All Days Pass' or 'Other'
import pandas as pd

# Load the ticket prices data for the specific festival into a DataFrame
ticket_prices = pd.read_csv("ticket_prices_rir.csv")

# Function to categorize ticket type for the specific festival
def categorize_ticket_type(ticket_name):
    ticket_name_lower = ticket_name.lower()
    
    if any(word in ticket_name for word in ['ENTRADA', 'DIÁRIO', 'JULY', 'AUGUST' 'August']) and 'VIP' not in ticket_name:
        return 'Daily'
    elif any(word in ticket_name for word in ['2 DIAS', 'ABONO', 'FIM DE SEMANA', 'PASS WITHOUT']):
        return 'Two Days Pass'
    elif any(word in ticket_name for word in ['3 DIAS', 'Pass', 'GENERAL PASS WITHOUT', 'PASS WITHOUT', 'Full']):
        return 'All Days Pass'
    else:
        return 'Other'

# Add a new column indicating ticket category for the specific festival
ticket_prices['Ticket Category'] = ticket_prices['Tipo de Bilhete'].apply(categorize_ticket_type)

# Save the DataFrame with the modified column to a new CSV file
ticket_prices.to_csv("ticket_prices_rir_categorized.csv", index=False)

print(ticket_prices)
```

```python
# Adding a 'Festival' column to each CSV file with the name of the festival the prices refer to
# merge all of them and save them in a CSV file called all_festivals_ticket_prices.csv

import pandas as pd

df_alive = pd.read_csv('ticket_prices_alive_categorized.csv')
df_alive['Festival'] = 'NOS Alive'

df_rir = pd.read_csv('ticket_prices_rir_categorized.csv')
df_rir['Festival'] = 'Rock in Rio'

df_sbsr = pd.read_csv('ticket_prices_sbsr_categorized.csv')
df_sbsr['Festival'] = 'Super Bock Super Rock'

df_sudoeste = pd.read_csv('ticket_prices_sudoeste_categorized.csv')
df_sudoeste['Festival'] = 'MEO Sudoeste'

df_vodafone = pd.read_csv('ticket_prices_vodafone_categorized.csv')
df_vodafone['Festival'] = 'Vodafone Paredes de Coura'

merged_df = pd.concat([df_alive, df_rir, df_sbsr, df_sudoeste, df_vodafone], ignore_index=True)

merged_df.to_csv('all_festivals_ticket_prices.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('all_festivals_ticket_prices.csv')

# Filter the dataframe to include only the rows in the 'Ticket Category' column with the categories of interest
categories_of_interest = ['Daily', 'Two Days Pass', 'All Days Pass']
df_filtered = df[df['Ticket Category'].isin(categories_of_interest)]

# Because there are duplicate entries (like several rows for Daily Tickets with the same value)
# we keep the mean of the 'Preço' column for each ticket category and for each festival
df_aggregated = df_filtered.groupby(['Festival', 'Ticket Category'])['Preço'].mean().reset_index()

# Sort the DataFrame by 'Festival' and 'Ticket Category'
df_sorted = df_aggregated.sort_values(by=['Festival', 'Ticket Category'])

# Define the order of ticket categories
category_order = ['Daily', 'Two Days Pass', 'All Days Pass']

# Pivot the sorted DataFrame to reshape it so that the first column is the 'Festival'
df_pivot = df_sorted.pivot(index='Festival', columns='Ticket Category', values='Preço')[category_order]

# To order from highest to lowest, a column with the total (sum of ticket prices for each festival) is created
df_pivot['Total'] = df_pivot.sum(axis=1)
df_pivot_sorted = df_pivot.sort_values(by='Total', ascending=False)    # sort the df_pivot by the 'Total' column
df_pivot_sorted.drop(columns=['Total'], inplace=True)   # and then it's deleted

palette = sns.color_palette('pastel')

colors = [palette[0], palette[3], palette[2]]

# The bar graph
plt.figure(figsize=(8, 6))
df_pivot_sorted.plot(kind='bar', stacked=True, color=colors, ax=plt.gca())
plt.title('Aggregated Ticket Prices per Festival')
plt.xlabel('Festival')
plt.ylabel('Price')
plt.xticks(rotation=30, ha='right')
plt.legend(title='Ticket Category')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('all_festivals_ticket_prices.csv')

# Filter the dataframe to include only the rows in the 'Ticket Category' column with the categories of interest
categories_of_interest = ['Daily', 'Two Days Pass', 'All Days Pass']
df_filtered = df[df['Ticket Category'].isin(categories_of_interest)]

plt.figure(figsize=(12, 8))

palette = sns.color_palette('pastel')
colors = [palette[0], palette[3], palette[2]]

# Define the order of ticket categories
category_order = ['Daily', 'Two Days Pass', 'All Days Pass']

# Plot three separate bars for each ticket category for all the festivals
sns.barplot(data=df_filtered, x='Festival', y='Preço', hue='Ticket Category', palette=colors, order=df['Festival'].unique(), hue_order=category_order)

plt.title('Ticket Prices per Festival and Category')
plt.xlabel('Festival')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend(title='Ticket Category')
plt.tight_layout()
plt.show()
```
