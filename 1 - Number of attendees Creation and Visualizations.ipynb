```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# To use the colors in pastel
palette = sns.color_palette('pastel')

# Festivals attendance data from online news
data_over_time = {
    'Year': [2018, 2019, 2020, 2021, 2022, 2023],
    'NOS Alive': [165000, 153000, 0, 0, 210000, 165000],
    'Rock in Rio': [278000, 0, 0, 0, 287000, 0],
    'MEO Sudoeste': [147000, 157000,0, 0, 0, 85000],
    'Super Bock Super Rock': [0, 55000, 0, 0, 45000, 58000],
    'Vodafone Paredes de Coura': [0, 0, 0, 0, 115000, 80000]
}

df = pd.DataFrame(data_over_time)

# Plot the graph 
plt.figure(figsize=(10, 6))
for column, color in zip(df.columns[1:], palette):
    plt.plot(df['Year'], df[column], marker='o', label=column, color=color)

plt.title('Number of Attendees per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attendees')
plt.xticks(df['Year'])
plt.legend()
plt.grid(False)  
plt.gca().set_facecolor('white') 
plt.tight_layout()
plt.show()
```
